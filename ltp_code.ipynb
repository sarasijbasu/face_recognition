```python
import os
import cv2
import math
import random
import numpy as np
import datetime as dt
from collections import deque
import matplotlib.pyplot as plt
import glob
main_path = os.getcwd()+"\\lfw_funneled"
labels = []
```

```python
def get_pixel(img, center, x, y):
    
    new_value = 0
    param = 10
    
    try:
        # If local neighbourhood pixel
        # value is greater than or equal
        # to center pixel values then
        # set it to 1
        if img[x][y] >= center+param:
            new_value = 1
        elif img[x][y] <= center-param:
            new_value = -1
        
            
    except:
        # Exception is required when
        # neighbourhood value of a center
        # pixel value is null i.e. values
        # present at boundaries.
        pass
    
    return new_value

# Function for calculating ltp
def ltp_calculated_pixel(img, x, y):

    center = img[x][y]
    val_ar = []
    
    # top_left
    val_ar.append(get_pixel(img, center, x-1, y-1))
    
    # top
    val_ar.append(get_pixel(img, center, x-1, y))
    
    # top_right
    val_ar.append(get_pixel(img, center, x-1, y + 1))
    
    # right
    val_ar.append(get_pixel(img, center, x, y + 1))
    
    # bottom_right
    val_ar.append(get_pixel(img, center, x + 1, y + 1))
    
    # bottom
    val_ar.append(get_pixel(img, center, x + 1, y))
    
    # bottom_left
    val_ar.append(get_pixel(img, center, x + 1, y-1))
    
    # left
    val_ar.append(get_pixel(img, center, x, y-1))
    
    # Now, we need to convert binary
    # values to decimal
    power_val = [1, 2, 4, 8, 16, 32, 64, 128]
    val = 0
    
    for i in range(len(val_ar)):
        val += val_ar[i] * power_val[i]
        
    return val
```

```python
labels = []
for directory_path in glob.glob(glob.glob(os.getcwd()+"/*")[2]+"/*"):
        labels.append(directory_path.split("\\")[-1])
for p in range (0,5760):
        current_dir = os.getcwd()+"\\lfw_funneled\\" + labels[p]
        for images in glob.glob(current_dir+"\\*.jpg"):
            img_bgr = cv2.imread(images,1)
            height, width, _ = img_bgr.shape
            img_gray = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2GRAY)
            img_ltp = np.zeros((height, width),np.uint8)

            for i in range(0, height):
                for j in range(0, width):
                    img_ltp[i, j] = ltp_calculated_pixel(img_gray, i, j)

            cv2.imwrite(images,img_ltp)
            print("Done upto "+labels[p])
```
