```python
import cv2
import numpy as np
import os
import pathlib
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
from tqdm import tqdm
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
## Define the main directory and list of all subdirectories
dir_images = "/content/drive/MyDrive/Classroom/orl_kamal/"
sub_dir = ["s1", "s2", "s3", "s4", "s5", "s6", "s7", "s8", "s9", "s10",
           "s11", "s12", "s13", "s14", "s15", "s16", "s17", "s18", "s19", "s20", 
           "s21", "s22", "s23", "s24", "s25", "s26", "s27", "s28", "s29", "s30", 
           "s31", "s32", "s33", "s34", "s35", "s36", "s37", "s38", "s39", "s40", 
           ]

for dirs in tqdm(sub_dir):    
    ## Define the read and write image paths
    read_img_path =  dir_images + dirs + "/"
    write_img_path =  "/content/drive/MyDrive/Classroom/orl_HOG/" + dirs + "_HOG/"
    pathlib.Path(write_img_path).mkdir(parents=True, exist_ok=True)
    imgs = os.listdir(read_img_path)

    ## Reading all images in the directory
    for imgnm in imgs:
        image = cv2.imread(os.path.join(read_img_path, imgnm))

        cv2_imshow(image)

        fd, hog_image = hog(image, orientations=7, pixels_per_cell=(4, 4), cells_per_block=(2, 2), visualize=True)
        cv2_imshow(hog_image)
        
        cv2.imwrite(write_img_path + "hog_" + imgnm, hog_image)
```
