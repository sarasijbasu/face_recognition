```python
import cv2
import glob
import os
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def median_thresholding(img):
    dummy = np.zeros((3,3),dtype = np.uint8)
    median = np.median(img.flatten())
    for i in range(3):
         for j in range(3):
            if img[i][j] >=median:
                 dummy[i][j] = 0
            else:
                dummy[i][j] = 255
    return dummy
```

```python
img = cv2.imread("C:\\Users\\saras\\Desktop\\image_pattern\\S010_004_00000017.png",0)
```

```python
for i in range(0,height,3):
    for j in range(0,width,3):
        container = median_thresholding(img[i:i+3,j:j+3])
        converted[i:i+3,j:j+3] = converted[i:i+3,j:j+3]+container
```
